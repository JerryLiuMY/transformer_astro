```python
def plot_count(catalog):
    fig, ax = plt.subplots(1, 1, figsize=(11, 7))
    sns.countplot(catalog['Class'])
```

```python
def plot_N(catalog):
    fig, ax = plt.subplots(1, 1, figsize=(11, 7))
    sns.distplot(catalog['N'], bins =25, hist=True, kde=False, ax=ax)
    ax_ = ax.twinx(); ax_.set_yticks([])
    sns.distplot(catalog['N'], hist=False, ax=ax_, label='All')
```

```python
def plot_N_group(catalog):
    fig, ax = plt.subplots(1, 1, figsize=(11, 7))
    for _ in list(set(catalog['Class'])): 
        sns.distplot(catalog[catalog['Class'] == _]['N'], bins =25, hist=False, kde=True, ax=ax, label=_)
```

```python
plot_count(macho_catalog)
plot_N(macho_catalog)
plot_N_group(macho_catalog)
```

```python
plot_count(asas_catalog)
plot_N(asas_catalog)
plot_N_group(asas_catalog)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, TimeDistributed
from keras.utils import to_categorical
import numpy as np
```

```python
def train_generator():
    while True:
        sequence_length = np.random.randint(10, 100)
        x_train = np.random.random((1000, sequence_length, 5))
        y_train = x_train[:, :, 0]
        for i in range(1, 5):
            y_train[:, i:] += x_train[:, :-i, i]
        y_train = to_categorical(y_train > 2.5)
        yield x_train, y_train
```
